```python
import pandas as pd
import numpy as np
```

```python
f = "../data/GI50/DTP_NCI60_RAW.csv"
df = pd.read_csv(f, header=8, na_values="na")
del f
```

# Reading in GI50 file

```python
df = df.groupby("NSC #").agg("mean")
df
```

```python
index = df.index
df = df.iloc[:, 3:]
df.index = index
del index
df
```

# Reading in Drug Fingerprints Data

```python
drug_f = "../data/drugs/fingerprints.csv"
drug_df = pd.read_csv(drug_f)
del drug_f
drug_df.index = drug_df["NSC #"]
drug_df = drug_df.iloc[:, 1:]
drug_df
```

removing drugs from GI50 data that do not have an associated fingerprint

```python
np.all(np.isin(drug_df.index, df.index))
```

```python
identified_drugs = np.intersect1d(drug_df.index, df.index)
df = df.loc[identified_drugs]
del identified_drugs
df
```

# Reading in Genetic Data

```python
genetic_f = "../data/cell_line/RNA__Affy_HuEx_1.0_GCRMA.txt.txt"
genetic_df = pd.read_csv(genetic_f, sep="\t", na_values="-")
del genetic_f
genetic_df.head(10)
```

np.any(np.unique(genetic_df.index, return_counts=True)[1] > 1)

```python
genetic_df = genetic_df.iloc[:, 7:]
genetic_df = genetic_df.dropna(axis=1, how="all")
genetic_df
```

```python
cell_lines = genetic_df.columns
np.all(np.isin(cell_lines, df.columns))
cell_lines[~np.isin(cell_lines, df.columns)]
```

```python
df = df[cell_lines]
del cell_lines
df
```

# Sampling Data

randomly selecting 1000 drugs for experiment where are columns are full

```python
df = df.dropna(axis=0)
df = df.sample(n=100, random_state=36, axis=0)
```

# Melting Data

```python
df["NSC #"] = df.index
df = pd.melt(df, id_vars=["NSC #"], value_vars=df.columns)
df = df.rename(columns={"variable":"Cell Line", "value":"GI50 Value"})
df
```

# Appending Drug Fingerprints

```python
drug_df = drug_df.add_prefix("drug_fingerprint_")
df = pd.merge(df, drug_df, left_on="NSC #", right_on=drug_df.index)
del drug_df
df
```

```python
np.all(df.sum(axis=0) != 0)
```

```python
df = df.loc[:, df.sum(axis=0) != 0]
df
```

# Appending Genetic Features

first, need to transpose the genetic_df matrix

```python
genetic_df = genetic_df.transpose()
genetic_df = genetic_df.add_prefix("genetic_expr_")
genetic_df
```

now I can merge the tables

```python
df = pd.merge(df, genetic_df, left_on="Cell Line", right_on=genetic_df.index)
df
```

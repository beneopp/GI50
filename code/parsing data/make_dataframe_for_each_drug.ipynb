```python
import pandas as pd
import numpy as np
```

```python
fingerprint_df = pd.read_csv("../data/drugs/fingerprints.csv")

index = fingerprint_df["NSC #"]

fingerprint_df = fingerprint_df.iloc[:, 1:]
fingerprint_df.inde = index

fingerprint_df = fingerprint_df.add_prefix("drug_fingerprint_")

fingerprint_df
```

```python
f = "../data/GI50/DTP_NCI60_RAW.csv"
df = pd.read_csv(f, header=8, na_values="na")
df = df.groupby("NSC #").agg("mean")
df
```

```python
df = pd.merge(df, fingerprint_df, left_on=df.index, right_on=index)

index = df.index

df = df.iloc[:, 4:]
df.index = index
df.index.name = "NSC #"

df
```

```python
df.to_csv("../data/machine_learning_input.csv")
```

```python
import pandas as pd
import numpy as np
```

```python
f = "../data/output/DTP_NCI60_RAW.csv"
df = pd.read_csv(f, header=8, na_values="na")
df2 = df.iloc[:, 9:]
df2.index = df["NSC #"]

df = df2
del df2
```

```python
df
```

```python
drugs = np.unique(df.index)
```

```python
cell_lines = df.columns
cancer_types = np.unique([cell_line.split(":")[0] for cell_line in cell_lines])
```

```python
drugs = np.unique(df.index)
```

```python
data = np.zeros((len(drugs), len(cancer_types)))
stds = pd.DataFrame(data, index=drugs, columns=cancer_types)
stds.index.name = "NSC #"
stds.columns.name = "Cancer Type"
```

```python
stds
```

```python
type_to_line = dict()

for cell_line in cell_lines:
    
    cancer_type = cell_line.split(":")[0]
    
    if cancer_type in type_to_line:
        type_to_line[cancer_type].append(cell_line)
        
    else:
        type_to_line[cancer_type] = [cell_line]
```

```python
for drug in drugs:
    
    for cancer_type in cancer_types:
        
        cell_lines_in_subset = type_to_line[cancer_type]
        
        subset_df = df.loc[drug, cell_lines_in_subset]
        
        array = subset_df.to_numpy(dtype=float).flatten()
        
        try:
            std = np.std(array, where=~np.isnan(array))
        except:
            print("hello")

        stds.loc[drug, cancer_type] = std
        
stds
```

```python
stds.to_csv("../data/std_by_cancer_type.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = "../data/cell_line/RNA__Affy_HuEx_1.0_GCRMA.txt.txt"
df = pd.read_csv(f, sep='\t', na_values=["-", "Un"])
```

```python
df2 = df2.dropna(axis=1, how="all")
df2.shape
```

```python
from scipy.spatial.distance import pdist
```

```python
df3 = pd.DataFrame(np.zeros((58, 58)), index=df2.columns, columns=df2.columns)

for row in df2.columns:
    x = df2[row]
    for col in df2.columns:
        y = df2[col]
        df3.loc[row, col] = np.sqrt(np.sum((x - y) ** 2))
```

```python
df3
```

```python
f2 = "../data/cell_line/dists.csv"
df3.to_csv(f2)
```

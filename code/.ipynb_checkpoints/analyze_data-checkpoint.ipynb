```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
cell_miner_df = pd.read_csv("../data/output/DTP_NCI60_RAW.csv", header=8, na_values="na")
cell_miner_df
```

```python
stds = cell_miner_df.groupby("NSC #").agg("std").iloc[:,3:]
```

```python
stds
```

```python
stds.max()
```

the standard deviations for some data points appear to be too high

```python
plt.clf()
array = stds.to_numpy(dtype=float).flatten()
plt.hist(array)
plt.show()
```

I want to eliminate any values with standard deviations larger than 1

```python
large_std = stds > 1
large_std.shape
```

```python
cell_miner_mean_vals = cell_miner_df.groupby("NSC #").agg("mean").iloc[:, 3:]
cell_miner_mean_vals
```

```python
cell_miner_mean_vals[large_std] = np.nan
```

```python
cell_miner_mean_vals
```

```python
percentage_nans_cells = cell_miner_mean_vals.isna().sum(axis=0) / cell_miner_mean_vals.shape[0]

plt.clf()
plt.hist(percentage_nans_cells)
plt.title("Percentage of NAN Cell Values")
plt.show()
```

```python
percentage_nans_drugs = cell_miner_mean_vals.isna().sum(axis=1) / cell_miner_mean_vals.shape[1]

plt.clf()
plt.hist(percentage_nans_drugs)
plt.title("Percentage of NAN Drug Values")
plt.show()
```

```python
cell_miner_mean_vals.to_csv("../data/mean_GI50.csv")
```
